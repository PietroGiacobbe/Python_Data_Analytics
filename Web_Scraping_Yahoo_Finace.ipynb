```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = "https://finance.yahoo.com/crypto/"

r = requests.get(url)
html_content = r.text
```

```python


soup = BeautifulSoup(html_content, 'html.parser') 
print(soup)
```

```python
table = soup.find('table', {'class': 'W(100%)'})
crypto_data = {}
for tr in table.find_all('tr')[1:]:
    crypto_info = {}
    tds = tr.find_all('td')
    crypto_info['Name'] = tds[0].text.strip()
    crypto_info['Price (USD)'] = tds[1].text.strip()
    crypto_info['Change'] = tds[2].text.strip()
    crypto_info['Market Cap (USD)'] = tds[3].text.strip()
    crypto_info['Volume in Currency (24Hr)'] = tds[4].text.strip()
    crypto_info['Circulating Supply'] = tds[5].text.strip()
    crypto_info['Volume in Crypto (24Hr)'] = tds[6].text.strip()
    crypto_data[crypto_info['Name']] = crypto_info
df = pd.DataFrame.from_dict(crypto_data, orient='index')
print(df)
```
